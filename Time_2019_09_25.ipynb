```python
fs=8000 # Hz
ts = 1/fs
N = 8
n = np.arange(0,N,1)

x = np.sin(2*np.pi*1000*n*ts) + 0.5*np.sin(2*np.pi*2000*n*ts + 3*np.pi/8)

X=np.zeros(N,dtype=np.complex)

for m in range(N):
    for n in range(N):
        X[m] = X[m] + x[n]*(np.cos(2*np.pi*m*n/N)-1j*np.sin(2*np.pi*m*n/N))
m=np.arange(0,N)
f = m*fs/N
plt.figure()
plt.plot(f,np.abs(X),'o')
plt.xlabel('Frekvens (Hz)')
plt.ylabel('Magnitude')
plt.ylim(0,np.max(np.abs(X)))
plt.grid()
plt.show()
```
